```python
from typing import Dict


class Vertex:
    def __init__(self, key: int) -> None:
        self.id = key
        self.connect_to = {}

    def get_id(self) -> None:
        return self.id

    def get_connections(self) -> Dict:
        return self.connect_to

    def add_neighbor(self, nbr_id: int, weight: int = 0) -> None:
        self.connect_to[nbr_id] = weight

    def get_weight(self, nbr_id: int):
        return self.connect_to[nbr_id]

    def __str__(self) -> str:
        return f"This {self.id} and is connected to {[x.id for x in self.connect_to]}"


class Graph:
    def __init__(self) -> None:
        vert_list = {}
        num_of_vertices = 0

        def get_vert_list(self) -> Dict:
            return self.vert_list

        def add_vert(self, key: int) -> None:
            new_vert = Vertex(key)
            self.vert_list[key] = new_vert

        def get_vert(self, key: int) -> Vertex:
            return self.vert_list[key]

        def add_edge(self, fro: int, to: int, cost:int) -> None:
            if fro not in self.vert_list:
                self.vert_list[fro] = self.add_vert(fro)
            
            if to not in self.vert_list:
                self.vert_list[to] = self.add_vert(to)
            
            fro.add_neighbor(to, cost)

    

def build_graph(word_file: str) -> Graph:
    word_dict = {} # {variant: [word1, word2]} for e.g mate -> {_ate: [rate,late], m_te: [mate, mite]}
    graph = Graph()

    # read the file 
    opened_word_file = open(word_file, "r")
    
    # create buckets of words that differ by one letter
    for line in opened_word_file:
        print(f"line: {line}")

        word = line[:-1] # word = line

        print(f"word: {word}")

        # create variants of the word that differ by one letter 
        for i in range(len(word)): # -> _ate, m_te, ma_e, mat_
            bucket = f"{word[:i]}_{word[i+1:]}" # -> _ate
            

            # add word to it corresponding variant list
            if bucket in word_dict:
                word_dict[bucket].append(word)

            # add word as a variant in the word dictionary
            else:
                word_dict[bucket] = word

    for bucket in word_dict.keys(): # _ate , m_ate, ma_e, mat_
        for word1 in word_dict[bucket]: # mate, rate in _ate
            for word2 in word_dict[bucket]: # mate, rate in _ate
                if word1 != word2:
                    graph.add_edge(word1,word2)

    return graph

```
